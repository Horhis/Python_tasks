# Задача 3

Постройте subplot размером 2-2 (т.е. 4 графика). На первом subplot отобразите график в котором сложите оси y из 1 и в торого списка, у второго разность, у третьего произведение и у четвертого деление. Подпишите графики

```python
# Место для решения задачи
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.random.randint(10, size=(5))
y = np.random.randint(10, size=(5))
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 5))
x1 = [math.log10(a) for a in x]
y1 = [y1 * 5 for y1 in x1]
axes[0, 0].plot(x, (y + y1), 'b')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y+y1')
axes[0, 0].set_title('Сумма у', fontsize=18)

axes[0, 1].plot(x, (y - y1), 'r')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y - y1')
axes[0, 1].set_title('Разность y', fontsize=18)

axes[1, 0].plot(x, (y * y1), 'g')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y * y1')
axes[1, 0].set_title('Произведение y', fontsize=18)

axes[1, 1].plot(x, (y / y1), 'y')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('y / y1')
axes[1, 1].set_title('Частное y', fontsize=18)
```
