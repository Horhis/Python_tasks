*While - цикл, который исполняется до тех пор пока условие верно* 

```python
print(1)
print(2)
print(3)
print(4)
print(5)
```

```python
# бесконечный цикл
number = 0

while number!=10:
    print(number)
```

```python
# ввод счетчика i - считает кол операций
number = 0
i = 0

while i != 11:
    print(number)
    i += 1 # i = i + 1
    number += 1
```

```python
# number - это счетчик и переменная для вывода
number = 0

while number != 11:
    print(number)
    number += 1
```

```python
# Играем в дурака и за проигрышь отнимается 5 баллов

Sofia = 60
Timur = 34

while Sofia >= Timur:
    Sofia-=5
    print(Sofia)
```
